```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
import csv
from google.cloud import firestore

db = firestore.Client()

CSV_FILE = "event_judge_mapping.csv"

with open(CSV_FILE, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    for row in reader:
        judges = [
            {"order": i + 1, "name": name.strip()}
            for i, name in enumerate(row["judges"].split("|"))
            if name.strip()
        ]

        db.collection("eventJudgeMapping").document(row["\ufeffeventCode"]).set({
            "eventCode": row["\ufeffeventCode"],
            "group": row["group"],
            "displayName": row["displayName"],
            "eventRef": row["eventRef"],
            "expectedJudgeCount": len(judges),
            "judgeOrder": judges
        })

print("eventJudgeMapping uploaded")
```
